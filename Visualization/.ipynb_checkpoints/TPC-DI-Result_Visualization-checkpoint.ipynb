## Heatmap

```python
log10_data = np.log10(data)
log10_data
```

```python
fig,ax = plt.subplots(1,1,figsize = (10,10))
sns.heatmap(log10_data,annot=True)
fig.suptitle("Heatmap of Historical Load Time (Seconds, Scaled by Log10)",size=15)
fig.savefig("Heatmap.jpg",dpi = 200)
```

```python
data2 = data.sum(axis = 0)
```

```python
fig2,ax2 = plt.subplots(1,1,figsize = (10,10))
fig2.suptitle("Lineplot of Total Historical Load Time of Different Scale Factor",size = 15)

sns.lineplot(data = pd.DataFrame(data2),ax=ax2)
sns.scatterplot(data = pd.DataFrame(data2),ax=ax2,palette=['r'])
ax2.get_legend().remove()
ax2.text(x="Scale Factor 3",y=40,s="86.064")


ax2.text(x="Scale Factor 6",y=232,s="278.56")


ax2.text(x="Scale Factor 12",y=945,s="991.168")


ax2.text(x="Scale Factor 24",y=4144,s="4181.5")


ax2.set_ylabel("Seconds",size = 15)
fig2.savefig("Lineplot.jpg",dpi = 200)
```
